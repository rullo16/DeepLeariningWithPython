#Universal Workflow of Machine Learning

The universal workflow of machine learning is broadly structured in three parts:



1.   Define the task-Understand the problem domain and the business logic underlying the customer asked for. Collect a dataset, understand wht the data represents and choose how to measure success on the task.
2.   Develop a model-Prepare your data so that it can be processed by a machine learning model, select a model evaluation protocol and a simple baseline to beat, train a first model that has generalization power and that can overfit, and then regularize and tune your model until you achieve the best possible generalizatio performance.
3.   Deploy the model-Present your work to stakeholders, ship the model to a web server, a mobile app, a web page, or an embedded device, monitor the model's performance in the wild, and start collecting the dat you need to build the next-gen model.



##Define the task

###Frame the problem
Framing a machine learning problem usually involves detailed discussions with stakeholders. Here are the questions that should be on the top of your mind:



*   What will the input data be? What are you trying to predict? You can only learn to predict something if you have training data available: for example you can only learn to classify the sentiment of movie reviews if you have both movie reviews and sentiment annotations available. Data availability is usually the limiting factor at this stage. In many cases, you will have to resort to collecting and annotating new datasets.
*   What type of machine learning task are you facing? Binary class? Multiclass class? etc...
In some cases it may be that machine learning isn't even the best to make sense of the data, and you should use something else, such as plain old-school statistical analysis.

> * Photo search engine is multiclass, multilabel classification
* Spam detector is binary classification
* Music recommendation engine turns out to be better handled not via deep learning but via matrix factorization
* Credit card fraud detection project is a scalar regression task
* Click-through-rate prediction project is binary classification task
* Anomalous cookie detection is binary classification, but also require an object detection model to correctly crop out the cookies in raw images.
* Finding new archeological sites from satellite images is an image-similarity ranking task.
>

* What do existing solutions look like? 
* Are there particular constraint to deal with?

Once the research is done, you should know what your inputs will be, what your targets will be and what broad type of ML task the problem maps to.
Hypothesis making at this stage:
* Targets can be predicted given inputs
* Data available is sufficiently informative

Until you have a working model, these are only hypothesis.
Not all problems can be solved with machine learning.







#Prepare Data

##Vectorization
All inputs and targets in a Neural Network must typically be tensors of floating-point data. Whatever data you need to process you must turn it into tensors, this step is called data vectorization.

##Value Normalization
In general, it isn't safe to feed into a nerual network data that takes relatively large values or data that is heterogeneous. Doing so can trigger large gradient updates preventing the network from converging. To make learning easier your data should have the following characteristics:
* Take small values - Between 0-1 range
* Be homogeneous - All features should take values in roughly the same range.

Additionally, stricter normalization practice is common and can help, but isn't always necessary:
* Normalize each feature independently to have mean 0
* Normalize each feature independently to have standard deviation of 1

x -= x.mean(axis=0)
x /= x.std(axis=0)

###Handling Missing Values

What if a feature wasn't available for all samples?
You could just discard the feature entirely, but you don't have to.
* If it is a categorical feature, it's safe to create a new category that means "value missing". The model will automatically learn what this implies with respect to the targets.
* If the feature is numerical, avoid inputting arbitrary value like '0', because it may create discontinuity in the latent space formed by the features, making it harder for a model trained on it to generalize. Instead consider replacing the missing value with the average or median value for the feature in the dataset.
